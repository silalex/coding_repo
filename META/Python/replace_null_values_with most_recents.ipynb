## Fill in the Blanks

```
Given an array containing null values fill in the null values with most recent non-null value in the array.

Signature
integer[] returnFilledArray(integer[] input_lst)

Input
Array of integers and/or null values

Output
Array of integers and/or null values

Examples
input:  [1,null,2,3,null,null,5,null]
output: [1, 1, 2, 3, 3, 3, 5, 5]

input: [null,8,null]
output: [null, 8, 8]

input: [1,null,2]
output: [1,1,2]

input: [5,null,null]
output: [5,5,5]

input: [1,null,2,3,null,null,5,null]
output: [1,1,2,3,3,3,5,5]
```

````python
# Given an array containing null values fill in the null values with most recent non-null value in the array.

def fill_in_the_blanks(input_lst: list) -> list:
  # Write your code here
  output_lst = []
  last_non_null = None

  for i in input_lst:
    if i is not None:
      output_lst.append(i)
      last_non_null = i
    else:
      output_lst.append(last_non_null)

  return output_lst

######## Test cases
print(fill_in_the_blanks([1, None, 2, None, 3]))  # Output: [1, 1, 2, 2, 3]
print(fill_in_the_blanks([None, 1, None, 2, None, 3, None]))  # Output: [1, 1, 2, 2, 3, 3]
print(fill_in_the_blanks([1, 2, None, None, 3, None, 4]))  # Output: [1, 2, 2, 2, 3, 3, 4]
````
