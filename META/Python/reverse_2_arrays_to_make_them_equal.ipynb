## Reverse to Make Equal

```
Given two arrays A and B of length N, determine if there is a way to make A equal to B by reversing any subarrays from array B any number of times.

Signature
bool areTheyEqual(int[] arr_a, int[] arr_b)

Input
All integers in array are in the range [0, 1,000,000,000].

Output
Return true if B can be made equal to A, return false otherwise.

Example
A = [1, 2, 3, 4]
B = [1, 4, 3, 2]
output = true

After reversing the subarray of B from indices 1 to 3, array B will equal array A.
```

````python
# Solution #1: use "Counter" sub-Class
from collections import Counter

def are_they_equal(array_a: list, array_b: list) -> bool:
  if len(array_a) != len(array_b):
    return False

  return Counter(array_a) == Counter(array_b)

######### Test cases
print(are_they_equal([1, 2, 3, 4], [1, 4, 3, 2]))  # True
print(are_they_equal([1, 2, 3, 4], [1, 2, 3, 5]))  # False
print(are_they_equal([1, 2, 3, 4], [4, 3, 2, 1]))  # True
print(are_they_equal([1, 2, 3, 4], [4, 3, 2, 2]))  # False
print(are_they_equal([1, 2, 3, 4], [4, 3, 2, 1, 5]))  # False  
````

````python
# Solution #2: w/o "Counter" sub-Class
def are_they_equal_manual(A, B):
  if len(A) != len(B):
    return False

  freq = {}

  # Count elements in A
  for x in A:
    freq[x] = freq.get(x, 0) + 1

  # Subtract using elements in B
  for x in B:
    if x not in freq:
      return False
    freq[x] -= 1
    if freq[x] < 0:
      return False

  return True

######### Test cases
print(are_they_equal_manual([1, 2, 3, 4], [1, 4, 3, 2]))  # True
print(are_they_equal_manual([1, 2, 3, 4], [1, 2, 3, 5]))  # False
print(are_they_equal_manual([1, 2, 3, 4], [4, 3, 2, 1]))  # True
print(are_they_equal_manual([1, 2, 3, 4], [4, 3, 2, 2]))  # False
print(are_they_equal_manual([1, 2, 3, 4], [4, 3, 2, 1, 5]))  # False  
````
