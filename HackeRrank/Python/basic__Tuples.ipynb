## Learn about 'tuples' and compute hash(T)

#### Task
```
Given an integer, n, and n space-separated integers as input, create a tuple, t, of those  integers. Then compute and print the result of hash(t).

Note: hash() is one of the functions in the __builtins__ module, so it need not be imported.

Input Format

The first line contains an integer, n, denoting the number of elements in the tuple.
The second line contains n space-separated integers describing the elements in tuple t.

Output Format:
Print the result of hash(t).

Sample Input:
2
1 2

Sample Output:
3713081631934410656
```

````python
# RUN the code in Python 2 ENV(!)
if __name__ == '__main__':
    n = int(raw_input())
    integer_list = map(int, raw_input().split())
    
    t = tuple(integer_list)
    print(hash(t))
````

````python
# Example 1: Hashing an immutable tuple of integers
my_tuple = (1, 2, 3)
hash_value = hash(my_tuple)
print(f"The hash of {my_tuple} is: {hash_value}")
# Output will be an integer, e.g., The hash of (1, 2, 3) is: -2022708474
````
