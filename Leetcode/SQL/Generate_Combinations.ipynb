## Task 1: filter out combinations w/ 3+ matches

### How it works
```
 - set(combo) & exclude_set → finds common numbers
 - len(...) → counts how many match
 - matches < 3 → keeps combinations with 0, 1, or 2 matches
 - combinations with 3, 4, or 5 matches are excluded
```
### Example (excluded)
```
(1, 5, 10, 12, 18)  → 5 matches ❌
(1, 5, 10, 12, 7)   → 4 matches ❌
(1, 5, 10, 7, 21)   → 3 matches ❌
```

````python
from itertools import combinations

numbers = [1, 5, 7, 10, 12, 18, 21]
exclude_set = {1, 5, 10, 12, 18}

result = []

match_5, match_4, match_3 = 0, 0, 0

for combo in combinations(numbers, 5):
    matches = len(set(combo) & exclude_set)
    if matches < 3:
        result.append(combo)
    elif matches == 3:
        match_3 += 1
    elif matches == 4:
        match_4 += 1
    elif matches == 5:
        match_5 += 1

print("Total valid combinations:", len(result))
print("Combinations with 5 matches:", match_5)
print("Combinations with 4 matches:", match_4)
print("Combinations with 3 matches:", match_3)

for r in result:
    print(r)
````
