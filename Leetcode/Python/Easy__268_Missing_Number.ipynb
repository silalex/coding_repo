## 268. Missing Number - Sorted Array [0 ... n]
```
Given an array nums containing n distinct numbers in the range [0, n], return the only number in the range that is missing from the array.

Example 1:
Input: nums = [3,0,1]
Output: 2

Explanation:
n = 3 since there are 3 numbers, so all numbers are in the range [0,3]. 2 is the missing number in the range since it does not appear in nums.

Example 2:
Input: nums = [0,1]
Output: 2

Explanation:
n = 2 since there are 2 numbers, so all numbers are in the range [0,2]. 2 is the missing number in the range since it does not appear in nums.

Example 3:
Input: nums = [9,6,4,2,3,5,7,0,1]
Output: 8

Explanation:
n = 9 since there are 9 numbers, so all numbers are in the range [0,9]. 8 is the missing number in the range since it does not appear in nums.
```

````python
# Solution #1: Gauss Formula (Sum of first n numbers)
from typing import List

def missingNumber(nums: List[int]) -> int:
    n = len(nums)
    expected_sum = n * (n + 1) // 2   # "ideal" List sum (if no gaps)
    actual_sum = sum(nums)
    return expected_sum - actual_sum

# Test case
print(missingNumber([3, 0, 1, 2, 5, 6]))  # Output: 2
````

````python
# Solution #2: Using XOR operation
# Runtime 0 ms / Beats 100.00%
# Memory 20.58 MB / Beats 21.45%
def missingNumberXOR(nums: List[int]) -> int:
    result = len(nums)
    for i in range(result):
        result ^= i ^ nums[i]
    return result

# Test case
print(missingNumberXOR([3, 0, 1, 2, 5, 6]))  # Output: 4
````

````python
# Solution #3: using XOR for Sorted List [1 ... n+1]
def missing_1_to_nplus1(nums):
    n = len(nums)
    res = n + 1
    for i, num in enumerate(nums, start=1):
        res ^= i ^ num
    return res

print(missing_1_to_nplus1([1, 2, 4, 5, 6]))  # Output: 3
````
