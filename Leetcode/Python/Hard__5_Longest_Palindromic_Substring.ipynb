## 5. Longest Palindromic Substring
```
Given a string s, return the longest palindromic substring in s.

Example 1:
Input: s = "babad"
Output: "bab"
Explanation: "aba" is also a valid answer.

Example 2:
Input: s = "cbbd"
Output: "bb"
```

````python
# Solution
# Runtime  243 ms / Beats 80.86%
# Memory 19.40 MB / Beats 54.51%

def longestPalindrome(s: str) -> str:
    if not s:
        return ""

    start, end = 0, 0

    def expand(left, right):
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left -= 1
            right += 1
        return left + 1, right - 1

    for i in range(len(s)):
        # odd palindrome
        l1, r1 = expand(i, i)

        # even palindrome
        l2, r2 = expand(i, i+1)

        if r1 - l1 > end - start:
            start, end = l1, r1

        if r2 - l2 > end - start:
            start, end = l2, r2

    return s[start:end+1]

# Test cases
print(longestPalindrome("babad"))  # Output: "aba" or "bab"
print(longestPalindrome("cbbd"))   # Output: "bb"
print(longestPalindrome("a"))      # Output: "a"
print(longestPalindrome("ac"))     # Output: "a" or "c"
````
