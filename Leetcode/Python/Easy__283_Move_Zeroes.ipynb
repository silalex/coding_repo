## 283. Move Zeroes
```
Given an integer array nums, move all 0's to the end of it while maintaining the relative order of the non-zero elements.

Note that you must do this in-place without making a copy of the array.

Example 1:
Input: nums = [0,1,0,3,12]
Output: [1,3,12,0,0]

Example 2:
Input: nums = [0]
Output: [0]
```

````python
# Solution #1
# Runtime    6 ms / Beats 47.22%
# Memory 20.54 MB / Beats 53.21%

def moveZeroes(nums: list[int]) -> None:
    """
    Do not return anything, modify nums in-place instead.
    """
    nonzeros = 0

    for i in range(len(nums)):
        if nums[i] != 0:
            nums[nonzeros] = nums[i]
            nonzeros += 1

    for j in range(nonzeros, len(nums)):
        nums[j] = 0

# Example usage:
nums = [0, 1, 0, 3, 12]
moveZeroes(nums)
print(nums)  # Output: [1, 3, 12, 0, 0]
````

````python
# Solution #2
# Runtime    4 ms / Beats 61.40%
# Memory 20.58 MB / Beats 53.21%

def moveZeroes(nums: list[int]) -> None:
    """
    Do not return anything, modify nums in-place instead.
    """
    nonzeros = 0

    for i in range(len(nums)):
        if nums[i] != 0:
            nums[nonzeros], nums[i] = nums[i], nums[nonzeros]
            nonzeros += 1

# Example usage:
nums = [0, 1, 0, 3, 12]
moveZeroes(nums)
print(nums)  # Output: [1, 3, 12, 0, 0]
````
