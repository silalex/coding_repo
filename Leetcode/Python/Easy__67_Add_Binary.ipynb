## 67. Add Binary
```
Given two binary strings a and b, return their sum as a binary string.

Example 1:
Input: a = "11", b = "1"
Output: "100"

Example 2:
Input: a = "1010", b = "1011"
Output: "10101"
```

````python
# Solution
def addBinary(self, a: str, b: str) -> str:
        # Convert binary strings to integers, sum them, convert back to binary
        return bin(int(a, 2) + int(b, 2))[2:]  # [2:] removes the '0b' prefix

# Test cases
print(addBinary(None, "11", "1"))  # Output: "100"
print(addBinary(None, "1010", "1011"))  # Output: "10101"
print(addBinary(None, "0", "0"))  # Output: "0"
print(addBinary(None, "1111", "1111"))  # Output: "11110"
````
