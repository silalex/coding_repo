## 1470. Shuffle the Array
```
Given the array nums consisting of 2n elements in the form [x1,x2,...,xn,y1,y2,...,yn].

Return the array in the form [x1,y1,x2,y2,...,xn,yn].

Example 1:
Input: nums = [2,5,1,3,4,7], n = 3
Output: [2,3,5,4,1,7] 

Explanation: Since x1=2, x2=5, x3=1, y1=3, y2=4, y3=7 then the answer is [2,3,5,4,1,7].

Example 2:
Input: nums = [1,2,3,4,4,3,2,1], n = 4
Output: [1,4,2,3,3,2,4,1]

Example 3:
Input: nums = [1,1,2,2], n = 2
Output: [1,2,1,2]
```

````python
# solution #1
# Runtime   54 ms / Beats 94.22%
# Memory 19.30 MB / Beats 85.31%

def shuffle(nums: list[int], n: int) -> list[int]:
    res = []

    if not nums or len(nums) % 2 == 1:
        return res

    for i in range(n):
        res.append(nums[i])
        res.append(nums[i + n])

    return res

# Test cases
print(shuffle([2, 5, 1, 3, 4, 7], 3)) # [2, 3, 5, 4, 1, 7]
print(shuffle([1, 2, 3, 4, 4, 3, 2, 1], 4)) # [1, 4, 2, 3, 3, 2, 4, 1]
print(shuffle([1, 1, 2, 2], 2)) # [1, 2, 1, 2]
print(shuffle([1, 2, 3], 2)) # []
````
