## Longest Substring Without Repeating Characters

```
Given a string s, find the length of the longest substring without duplicate characters.

Example 1:

Input: s = "abcabcbb"
Output: 3
Explanation: The answer is "abc", with the length of 3. Note that "bca" and "cab" are also correct answers.

Example 2:

Input: s = "bbbbb"
Output: 1
Explanation: The answer is "b", with the length of 1.

Example 3:

Input: s = "pwwkew"
Output: 3
Explanation: The answer is "wke", with the length of 3.

Notice that the answer must be a substring, "pwke" is a subsequence and not a substring.
```

````python
# Given a string s, find the length of the longest substring without duplicate characters.

class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        char_index = {}  # Dictionary to store the last index of each character
        left = 0         # Left pointer of the sliding window
        max_len = 0      # Maximum length of substring found

        for right, ch in enumerate(s):
            if ch in char_index and char_index[ch] >= left:
                left = char_index[ch] + 1 # Move left pointer to the right of the last occurrence

            char_index[ch] = right # Update the last index of the character
            max_len = max(max_len, right - left + 1) # Update maximum length if needed

        print(char_index)  # Debug: print the character index mapping
        return max_len
    
# Test cases:
if __name__ == "__main__":
    solution = Solution()
    test_string = "abcabcbb"
    print(f"The length of the longest substring without duplicate characters in '{test_string}' is: {solution.lengthOfLongestSubstring(test_string)}")

````
