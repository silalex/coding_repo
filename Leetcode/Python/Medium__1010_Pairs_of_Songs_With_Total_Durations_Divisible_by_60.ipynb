## 1010. Pairs of Songs With Total Durations Divisible by 60
```
You are given a list of songs where the ith song has a duration of time[i] seconds.

Return the number of pairs of songs for which their total duration in seconds is divisible by 60. Formally, we want the number of indices i, j such that i < j with (time[i] + time[j]) % 60 == 0.

Example 1:
Input: time = [30,20,150,100,40]
Output: 3

Explanation: Three pairs have a total duration divisible by 60:
 - (time[0] = 30, time[2] = 150): total duration 180
 - (time[1] = 20, time[3] = 100): total duration 120
 - (time[1] = 20, time[4] = 40): total duration 60

Example 2:
Input: time = [60,60,60]
Output: 3

Explanation: All three pairs have a total duration of 120, which is divisible by 60.
```

````python
# Solution #1 (brute force) -> exeecution: Time Limit Exceeded(!)
class Solution:
    def numPairsDivisibleBy60(self, time):
        count = 0
        for i in range(len(time)):
            for j in range(i + 1, len(time)):
                if (time[i] + time[j]) % 60 == 0:
                    count += 1
        return count
    
# Test case
time = [30, 20, 150, 100, 40]
solution = Solution()
print(solution.numPairsDivisibleBy60(time))  # Output: 3
````

````python
# Solution #2 (using a hash map)
# Runtime: 35 ms / Beats 14.45%
# Memory: 23.16 MB / Beats 36.68%
class Solution:
    def numPairsDivisibleBy60(self, time):
        rem_cnt = {} # key: remainder, value: count of occurrences
        cnt = 0

        for t in time:
            rem = t % 60 # remainder when divided by 60 (eg 150 % 60 = 30)
            num = (60 - rem) % 60 # find required num until the sum is divisible by 60

            if num in rem_cnt:
                cnt += rem_cnt[num]
                print(f"[{t}] {num} for {rem} FOUND(!) (cnt {cnt})") # debug print to show found required num
            else:
                print(f"[{t}] {num} for {rem} NOT FOUND (cnt {cnt})") # debug print to show missing required num

            rem_cnt[rem] = rem_cnt.get(rem, 0) + 1

            print(rem_cnt) # debug print to show remainder counts
        
        return cnt # return total count of valid pairs

# Test case
time = [30, 20, 150, 100, 40, 80]
solution = Solution()
print(solution.numPairsDivisibleBy60(time))  
# Output: count = 5
# remainder_count: {30: 3, 20: 2, 40: 2}
````

````python
# Solution #3 (using a Modulo 60 frequency array)
# Runtime: 19 ms / Beats 63.73%
# Memory: 23.20 MB / Beats 28.97%
class Solution:
    def numPairsDivisibleBy60(self, time):
        rem_cnt = [0] * 60 # frequency array for remainders
        cnt = 0

        for t in time:
            rem = t % 60 # remainder when divided by 60
            num = (60 - rem) % 60 # required num to make sum divisible by 60 (complementary remainder)

            cnt += rem_cnt[num] # add count of required num directly

            rem_cnt[rem] += 1 # increment the count for this remainder
        
        return cnt # return total count of valid pairs
    
# Test case
time = [30, 20, 150, 100, 40]
solution = Solution()
print(solution.numPairsDivisibleBy60(time))  # Output: 3
````
