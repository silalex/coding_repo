## 1010. Pairs of Songs With Total Durations Divisible by 60
```
You are given a list of songs where the ith song has a duration of time[i] seconds.

Return the number of pairs of songs for which their total duration in seconds is divisible by 60. Formally, we want the number of indices i, j such that i < j with (time[i] + time[j]) % 60 == 0.

Example 1:
Input: time = [30,20,150,100,40]
Output: 3

Explanation: Three pairs have a total duration divisible by 60:
 - (time[0] = 30, time[2] = 150): total duration 180
 - (time[1] = 20, time[3] = 100): total duration 120
 - (time[1] = 20, time[4] = 40): total duration 60

Example 2:
Input: time = [60,60,60]
Output: 3

Explanation: All three pairs have a total duration of 120, which is divisible by 60.
```

````python
# Solution #1 (brute force) -> exeecution: Time Limit Exceeded(!)
class Solution:
    def numPairsDivisibleBy60(self, time):
        count = 0
        for i in range(len(time)):
            for j in range(i + 1, len(time)):
                if (time[i] + time[j]) % 60 == 0:
                    count += 1
        return count
    
# Test case
time = [30, 20, 150, 100, 40]
solution = Solution()
print(solution.numPairsDivisibleBy60(time))  # Output: 3
````

````python
# Solution #2 (using a hash map)
class Solution:
    def numPairsDivisibleBy60(self, time):
        remainder_count = {} # key: remainder, value: count of occurrences
        count = 0 # number of valid pairs
        
        for t in time: # iterate through each time duration
            remainder = t % 60 # remainder when divided by 60
            complement = (60 - remainder) % 60 # complement remainder to make sum divisible by 60
            
            if complement in remainder_count: # check if complement exists
                count += remainder_count[complement] # add the count of complement occurrences
            
            if remainder in remainder_count: # update the count of the current remainder
                remainder_count[remainder] += 1 # increment count
            else:
                remainder_count[remainder] = 1 # initialize count

            print(remainder_count) # debug print to show remainder counts
        
        return count # return total count of valid pairs
    
# Test case
time = [30, 20, 150, 100, 40, 80]
solution = Solution()
print(solution.numPairsDivisibleBy60(time))  
# Output: count = 5
# remainder_count: {30: 3, 20: 2, 40: 2}
````
