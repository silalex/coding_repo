### Pool Two Trips Within 60 Minutes
```
Problem

Uber wants to offer a “pool driver” feature where a driver can do two trips back-to-back within 1 hour total.

You are given an array durations, where:

durations[i] = trip duration (minutes)

trip_id = i (index is the id)

Return all pairs of trip_ids (i, j) such that:

i < j

durations[i] + durations[j] <= 60

Example
durations = [15, 35, 45, 10, 20]

Valid pairs:
(0,1) = 15+35=50
(0,2) = 15+45=60
(0,3) = 15+10=25
(0,4) = 15+20=35
(1,3) = 35+10=45
(1,4) = 35+20=55
(3,4) = 10+20=30

Output:
[(0,1),(0,2),(0,3),(0,4),(1,3),(1,4),(3,4)]
```
