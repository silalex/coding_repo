##  Longest Substring w/o Repeating Chars (Sliding Window)
```
Solution
Given a string s, find the length of the longest substring without duplicate characters.

Example 1:
Input: s = "abcabcbb"
Output: 3

Explanation: The answer is "abc", with the length of 3. Note that "bca" and "cab" are also correct answers.

Example 2:
Input: s = "bbbbb"
Output: 1

Explanation: The answer is "b", with the length of 1.

Example 3:
Input: s = "pwwkew"
Output: 3

Explanation: The answer is "wke", with the length of 3. Note that "kew" is also correct answers.

Notice that the answer must be a substring, "pwke" is a subsequence and not a substring.
```

````python
def length_of_the_longest_substring(self, s: str) -> int:
    substr = set()
    left = 0
    max_len = 0

    for right in range(len(s)):
        while s[right] in substr:
            substr.remove(s[left])
            left += 1

        substr.add(s[right])
        max_len = max(max_len, right - left + 1)

    return max_len

# Example usage:
if __name__ == "__main__":
    s = "abcadefbcbb"
    print(length_of_the_longest_substring(None, s))  # Output: 6
````
