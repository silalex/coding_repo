## 182. Duplicate Emails
```
Table: Person
+-------------+---------+
| Column Name | Type    |
+-------------+---------+
| id          | int     |
| email       | varchar |
+-------------+---------+
id is the primary key (column with unique values) for this table.
Each row of this table contains an email. The emails will not contain uppercase letters.

Write a solution to report all the duplicate emails. Note that it's guaranteed that the email field is not NULL.

Return the result table in any order.

The result format is in the following example.
```
#### Example
```
Input: 
Person table:
+----+---------+
| id | email   |
+----+---------+
| 1  | a@b.com |
| 2  | c@d.com |
| 3  | a@b.com |
+----+---------+

Output: 
+---------+
| Email   |
+---------+
| a@b.com |
+---------+

Explanation: a@b.com is repeated two times.
```

````python
### Solution #1
# Runtime: 293 ms / Beats 29.29%
# Memory: 67.38 MB / Beats 41.94%

import pandas as pd

def duplicate_emails(person: pd.DataFrame) -> pd.DataFrame:
    duplicates = (
        person
        .groupby("email", as_index=False)
        .size()
        .query("size > 1")
        [["email"]]
        .rename(columns={"email": "Email"})
    )
    return duplicates

# Test Case
data = [[1, 'a@b.com'], [2, 'c@d.com'], [3, 'a@b.com']]
person = pd.DataFrame(data, columns=['id', 'email']).astype({'id':'Int64', 'email':'object'})
print(duplicate_emails(person))
````

````python
## Solution #2:
# Runtime 273 ms & Beats 62.05%
# Memory 67.05 MB & Beats 74.02%
def duplicate_emails(person: pd.DataFrame) -> pd.DataFrame:
    return (
        person[person.duplicated("email", keep=False)]
        [["email"]]
        .drop_duplicates()
        .rename(columns={"email": "Email"})
    )

# Test Case
data = [[1, 'a@b.com'], [2, 'c@d.com'], [3, 'a@b.com']]
person = pd.DataFrame(data, columns=['id', 'email']).astype({'id':'Int64', 'email':'object'})
print(duplicate_emails(person))
````
