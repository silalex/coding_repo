### 1. Remove duplicates while preserving order

Task Desc: How do you remove duplicates from a Python list while preserving order?

```python
def remove_dups(lst):

    seen = set()
    result = []
    
    for x in lst:
        if x not in seen:
            seen.add(x)
            result.append(x)
    
    return result


if __name__ == "__main__":
    sample_list = [1, 2, 2, 3, 4, 4, 5]
    print("Original list:", sample_list)
    print("List after removing duplicates:", remove_dups(sample_list))
```
