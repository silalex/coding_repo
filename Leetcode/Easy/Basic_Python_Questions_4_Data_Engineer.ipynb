### 1. Remove duplicates while preserving order

Task Desc: How do you remove duplicates from a Python list while preserving order?

```python
def remove_dups(lst):

    seen = set()
    result = []
    
    for x in lst:
        if x not in seen:
            seen.add(x)
            result.append(x)
    
    return result


if __name__ == "__main__":
    sample_list = [1, 2, 2, 3, 4, 4, 5]
    print("Original list:", sample_list)
    print("List after removing duplicates:", remove_dups(sample_list))
```

### 2. Return only even numbers

Task Desc: Given a list of numbers, write code to return only even numbers.

Solution: by using List comprehension

```python
def get_evens(nums):
    evens = [n for n in nums if n % 2 == 0]
    return evens

if __name__ == "__main__":
    sample_nums = [1, 2, 3, 4, 5, 6]
    print("Even numbers:", get_evens(sample_nums))
```

### 3. Flatten list of lists

Task Desc: How do you flatten a list of lists in Python?

Solution: a double for-loop inside a list comprehension

```python
def list_flatten(nested):
    flat = [x for sub in nested for x in sub]
    return flat

if __name__ == "__main__":
    nested_list = [[1, 2], [3, 4], [5]]
    print("Flattened list:", list_flatten(nested_list))
```
