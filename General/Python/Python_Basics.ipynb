# 1. Lists

```python
[1,2,3] == [1,2,3]  # True (same elements in same order)
```

```python
[1,2,3] == [3,2,1] # False (different order)
```

```python
[1,2,3] is [1,2,3]  # False (different objects in memory)
```

```python
records = [
    frozenset(["red", "blue"]),
    frozenset(["blue", "red"]),
    frozenset(["green"])
]

unique = set(records)
print(unique)
# {frozenset({'red', 'blue'}), frozenset({'green'})}
# removes duplicates based on the frozenset's content, not the order of elements
```
