```python
[1,2,3] == [3,2,1] # False (different order)
```

## frozenset is an immutable version of a set.

Like sets, it contains unique, unordered, unchangeable elements.

Unlike sets, elements cannot be added or removed from a frozenset.

```python
records = [
    frozenset(["red", "blue"]),
    frozenset(["blue", "red"]),
    frozenset(["green"])
]

unique = set(records)
print(unique)
# {frozenset({'red', 'blue'}), frozenset({'green'})}
# removes duplicates based on the frozenset's content, not the order of elements
```

## List comprehension

List comprehensions provide a concise way to create lists.

### Rule of thumb:
 - List comprehension = “build a list”
 - Loop = “do a process”

If it starts feeling hard to read → switch back to a loop.

### Use list comprehension when:
 - Short and simple transformation/filter
 - You want clean, compact code
 - You’re building a new list directly


```python
nums = [1, 5, 9, 10]
squares = [x*x for x in nums if x > 0]
squares
```

### Use loops when:
 - Logic is complex or multi-step
 - You need debugging / logging / intermediate steps
 - You have side effects (DB, print, API calls)

```python
# If you want real logging (recommended for production)
import logging

logging.basicConfig(level=logging.INFO)

nums = [1, 5, 9, 10]

result = []
for x in nums:
    if x > 0:
        y = x * x
        logging.info("Processing %d -> %d", x, y)
        result.append(y)

result
```
