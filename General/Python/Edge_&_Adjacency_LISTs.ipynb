# Edge Lists

```python
# SET of TUPLEs to represent undirected edges (de-duplicated)
edges = {
    ("Alice", "Bob"),
    ("Alice", "Bob"), # this duplicate will be ignored in a SEL
    ("Bob", "Alice")  # same undirected edge, can normalize
}

print(edges)
```

```python
for edge in edges:
    print(sorted(edge))
```

```python
# de-duplication: normalize edges so that (A, B) and (B, A) are treated the same
# normalization trick (interview gold)
edges = {tuple(sorted(edge)) for edge in edges}
print(edges)
```

```python
# LIST of LISTs
edges_lst = [
    ["Alice", "Bob"],
    ["Alice", "Bob"], # this duplicate will NOT be ignored in a LIST
    ["Bob", "Alice"]  # same undirected edge, can normalize
]

print(edges_lst)
```

```python
# normalization trick (interview gold)
edges_lst = [list(edge) for edge in {tuple(sorted(e)) for e in edges_lst}]
print(edges_lst)
```

```python
# Converting Edge List to Adjacency List
friends = [
    ["Alice", "Bob"],
    ["Alice", "Charlie"],
    ["Bob", "David"]
]

friends_list = {}

# Build the adjacency list as a bidirectional graph
for f1, f2 in friends:
    if f1 not in friends_list:
        friends_list[f1] = []
    if f2 not in friends_list:
        friends_list[f2] = []
    friends_list[f1].append(f2)
    friends_list[f2].append(f1)

print(friends_list)
```
