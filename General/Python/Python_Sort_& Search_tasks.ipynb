## 1) Quick Sort
```
Quicksort is one of the fastest sorting algorithms.

The Quicksort algorithm takes an array of values, chooses one of the values as the 'pivot' element, and moves the other values so that lower values are on the left of the pivot element, and higher values are on the right of it.
```

````python
def quicksort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]   # middle element
    left = [x for x in arr if x < pivot]
    mid  = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quicksort(left) + mid + quicksort(right)


nums = [3, 6, 8, 10, 1, 2, 1]
print(quicksort(nums))
````

## 2) Binary Search
The Binary Search algorithm searches through a sorted array and returns the index of the value it searches for.

The Binary Search algorithm works by checking the value in the center of the array. If the target value is lower, the next value to check is in the center of the left half of the array. This way of searching means that the search area is always half of the previous search area, and this is why the Binary Search algorithm is so fast.

````python
from typing import List

def binarySearch(nums: List[int], target: int) -> int:
    left, right = 0, len(nums) - 1

    while left < right:
        mid = (left + right) // 2
        if nums[mid] < target:
            left = mid + 1
        else:
            right = mid
    # a conditional expression
    return left if nums[left] == target else -1

nums = [-1, 0, 3, 5, 9, 12]
target = 9
print(binarySearch(nums, target))  # Output: 4
````

## 3) sorted() vs sort()

### **sorted()** returns a new sorted list of the specified iterable object (list, tuple, dict?).
### Only(!) List objects have a **sort()** method that will sort the list in place(!) alphanumerically, ascending (by default).

However, all **sequences** are **iterables**, but not all **iterables** are **sequences** (**iterables** > **sequences**)

#### Sequence Types: list, tuple, range, str, bytes, bytearray (Indexed + Fixed Order)
#### Iterable Types: list, tuple, range, str, bytes, bytearray + dict, set, file, generator, zip/map (can be used in LOOP)


````python
# To sort a list by length, we can use the built-in len function.

a = ("Jenifer", "Sally", "Jane")
x = sorted(a, key=len)
print(x)
````

````python
# To sort a list by length, we can use the built-in len function in DESC.

a = ("Jenifer", "Sally", "Jane")
x = sorted(a, key=len, reverse = True)
print(x)
````

````python
# Sort the list by the number closest to 10:

def my_func(n):
    return abs(10 - n)

a = (5, 3, 1, 11, 2, 12, 17)
x = sorted(a, key=my_func)
print(x)
````

````python
# Sort strings alphabetically, ignoring case
names = ["Peter", "ann", "Jane"]

x = sorted(names, key=str.lower)
print(x)
````

````python
# Sort numbers by absolute value
nums = [-10, 3, -2, 7, -5]

x = sorted(nums, key=abs)
print(x)
````
