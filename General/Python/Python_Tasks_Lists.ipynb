### 3) How to remove None values from the List of INTs

```python
import math

raw_data = [56.2, float('NaN'), 51.7, 55.3, 52.5, float('NaN'), 47.8]
filtered_data = []

for value in raw_data:
    if not math.isnan(value):
        filtered_data.append(value)

print(filtered_data)
```

### 4) Find all duplicates in an array

```python
# Solution #1: Using a dictionary (recommended for interviews)
def find_duplicates_1(nums: list[int]) -> list[int]:
    counts = {}
    duplicates = []

    for num in nums:
        counts[num] = counts.get(num, 0) + 1

    for num, cnt in counts.items():
        if cnt > 1:
            duplicates.append(num)

    return duplicates

# Test Cases
find_duplicates_1([4, 2, 7, 2, 8, 4, 9, 4])
```

```python
# Solution #2: Using collections.Counter (most Pythonic)
from collections import Counter

def find_duplicates_2(nums: list[int]) -> list[int]:
    counts = Counter(nums)
    return [num for num, cnt in counts.items() if cnt > 1]

# Test Cases
find_duplicates_2([4, 2, 7, 2, 8, 4, 9, 4])
```

```python
# Solution #3: Using sets (single pass)
def find_duplicates_3(nums: list[int]) -> list[int]:
    seen = set()
    duplicates = set()

    for num in nums:
        if num in seen:
            duplicates.add(num)
        else:
            seen.add(num)

    return duplicates

# Test Cases
find_duplicates_3([4, 2, 7, 2, 8, 4, 9, 4])
```

### 5) Find the first non-repeating element in an array

```python
# Solution #1: Using Counter (recommended)
from collections import Counter

def first_non_repeating_1(nums: list[int]) -> int | None:
    counts = Counter(nums)

    for num in nums:
        if counts[num] == 1:
            return num
        
    return None

# test Case:
print(first_non_repeating_1([4, 5, 1, 2, 1, 2, 4]))
```

```python
# Solution #2: Using a dictionary
def first_non_repeating_2(nums: list[int]) -> int | None:
    counts = {}

    for num in nums:
        counts[num] = counts.get(num, 0) + 1

    for num in nums:
        if counts[num] == 1:
            return num

    return None

# test Case:
print(first_non_repeating_2([4, 5, 1, 2, 1, 2, 4]))
```

### 6) Find the intersection of two arrays

```python
# Solution #1: Using sets (recommended) - Returns the unique intersection.
def intersection(nums1: list[int], nums2: list[int]) -> list[int]:
    return list(set(nums1) & set(nums2))

# Test Cases
a = [1, 2, 2, 3, 4]
b = [2, 2, 4, 5]

print(intersection(a, b))
```

### 7) Group words that are anagrams

```python
# Solution #1: defaultdict + sorted() (Recommended) - Time = O(n × k log k) / Space = O(n × k)
from collections import defaultdict

def group_anagrams_1(words: list[str]) -> list[list[str]]:
    groups = defaultdict(list)

    for word in words:
        key = "".join(sorted(word))
        groups[key].append(word)

    return list(groups.values())

# Test Cases
words = ["eat", "tea", "tan", "ate", "nat", "bat"]
print(group_anagrams_1(words))

```

```python
# Solution #2: Regular dictionary
def group_anagrams_2(words: list[str]) -> list[list[str]]:
    groups = {}

    for word in words:
        key = "".join(sorted(word))

        if key not in groups:
            groups[key] = []

        groups[key].append(word)

    return list(groups.values())

# Test Cases
words = ["eat", "tea", "tan", "ate", "nat", "bat"]
print(group_anagrams_2(words))

```
