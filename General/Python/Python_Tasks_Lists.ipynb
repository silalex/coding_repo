## List basics
### Note: slicing returns a new list

```python
squares = [1, 4, 9, 16, 25]
print(squares[-3:]) # last 3 lements
print(squares[:-3]) # without last 3 elements
```

```python
# Concatenate two Lists
squares + [36, 49, 64, 81, 100]
```

```python
# count the number of times a value appears in a list
fruits = ['orange', 'apple', 'pear', 'banana', 'kiwi', 'apple', 'banana']
fruits.count('apple')
```

```python
# count the number of times a value appears in a list
numbers = [2, 4, 5, 2, 2, 0, 2]
numbers.count(2)
```

```python
# reverse a list
numbers.reverse()
numbers
```

```python
# sort a list
numbers.sort()
numbers
```
