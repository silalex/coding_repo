### Task #1: Find First Unique Character in a String

```python
# Solution #1
s = 'loveleetcode'

for i, ch in enumerate(s):
    if s.index(ch) == s.rindex(ch):
        print(s[i])
        break
```

```python
# Solution #2
s = 'loveleetcode'

for i, ch in enumerate(s):
    if ch not in s[i+1:]:
        print(s[i])
        break
```

```python
# Solution #3 (the same complexity as Solution #1)
s = 'loveleetcode'

for ch in s:
    if s.index(ch) == s.rindex(ch):
        print(ch)
        break
```

### Compare Solution #1 vs Solution #2

| Metric                               | Solution 1 (`index/rindex`) | Solution 2 (`slices + not in`) |
| ------------------------------------ | --------------------------- | ------------------------------ |
| Time per iteration                   | O(n) + O(n)                 | O(n) + O(n) + O(n) + O(n)      |
| Worst-case Time                      | O(n²)                       | O(n²)                          |
| Best-case Time                       | O(n)                        | O(n)                           |
| Auxiliary Space                      | O(1)                        | O(n)                           |
| Creates temporary strings            | No                          | Yes                            |
| Number of string scans per iteration | 2                           | Up to 4                        |
| Readability                          | High                        | Medium                         |
| Performance                          | Faster                      | Slower                         |
