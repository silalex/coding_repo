### Task #1: Find First Unique Character in a String

```python
# Solution #1
s = 'loveleetcode'

for i, ch in enumerate(s):
    if s.index(ch) == s.rindex(ch):
        print(s[i])
        break
```

```python
# Solution #2
s = 'loveleetcode'

for i, ch in enumerate(s):
    if ch not in s[i+1:]:
        print(s[i])
        break
```

```python
# Solution #3 (the same complexity as Solution #1)
s = 'loveleetcode'

for ch in s:
    if s.index(ch) == s.rindex(ch):
        print(ch)
        break
```

### Compare Solution #1 vs Solution #2

| Metric                               | Solution 1 (`index/rindex`) | Solution 2 (`slices + not in`) |
| ------------------------------------ | --------------------------- | ------------------------------ |
| Time per iteration                   | O(n) + O(n)                 | O(n) + O(n) + O(n) + O(n)      |
| Worst-case Time                      | O(n²)                       | O(n²)                          |
| Best-case Time                       | O(n)                        | O(n)                           |
| Auxiliary Space                      | O(1)                        | O(n)                           |
| Creates temporary strings            | No                          | Yes                            |
| Number of string scans per iteration | 2                           | Up to 4                        |
| Readability                          | High                        | Medium                         |
| Performance                          | Faster                      | Slower                         |


### Task #2: Palindrome Check

Given a string s, return True if it is a palindrome, otherwise False. Ignore case.

```python
# Solution #1 - simpliest (just Ignore case)
def is_palindrome(s: str) -> bool:
    s = s.lower()
    return s == s[::-1]

print(is_palindrome("Level"))
```

```python
# Solution #2: with removing all not alphanum chars + lowercase
def is_palindrome_2(s: str) -> bool:
    res = ''.join(ch.lower() for ch in s if ch.isalnum())
    return res == res[::-1]

print(is_palindrome("_Level"))
print(is_palindrome_2("_Level"))
```

### Task #3: Count Character Frequency
Given a string s, return a dictionary mapping character → count.

```python
# Solution #1: manual counting
def count_char_freq(s: str) -> dict:
    freq = {}
    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1
    return freq

print(count_char_freq("hello world"))
```

```python
# Solution #2: by using collection
from collections import Counter

def count_char_freq_2(s: str) -> dict:
    return dict(Counter(s))

print(count_char_freq_2("hello world"))
```

### Performance comparison

| Aspect                   | Solution #1: Manual counting                          | Solution #2: `collections.Counter`          |
| ------------------------ | ----------------------------------------------------- | ------------------------------------------- |
| Time complexity          | O(n)                                                  | O(n)                                        |
| Space complexity         | O(k) (unique chars)                                   | O(k)                                        |
| Constant factors (speed) | Higher overhead (Python-level loop + dict operations) | Lower overhead (optimized C implementation) |
| Implementation           | Pure Python loop                                      | C-optimized library                         |
| Readability              | Medium                                                | High                                        |
| Maintainability          | Medium                                                | High                                        |
| Performance (real-world) | Slower                                                | Faster                                      |
| Memory efficiency        | Slightly more overhead                                | More optimized internals                    |
| Best use case            | Learning / custom logic                               | Production / standard frequency counting    |


### Task #4: Longest Substring Without Repeating Characters

Return the length of the longest substring without repeating characters.

Note: This is a classic sliding window problem.

```python
# Solution #1: using sliding window technique
def length_of_longest_substring(s: str) -> int:
    char_set = set()
    left = 0
    max_length = 0

    for right in range(len(s)):
        while s[right] in char_set:
            char_set.remove(s[left])
            left += 1
        char_set.add(s[right])
        max_length = max(max_length, right - left + 1)

    return max_length

# Test Cases
print(length_of_longest_substring("abcabcbb"))  # Output: 3
print(length_of_longest_substring("bbbbb"))     # Output: 1
print(length_of_longest_substring("pwwekaeew")) # Output: 4
print(length_of_longest_substring(""))          # Output: 0
```

```python
# Solution #2: using dictionary to store last seen index of characters
def length_of_longest_substring_2(s: str) -> int:
    char_index = {}
    left = 0
    max_length = 0

    for right in range(len(s)):
        if s[right] in char_index and char_index[s[right]] >= left:
            left = char_index[s[right]] + 1
        char_index[s[right]] = right
        max_length = max(max_length, right - left + 1)

    return max_length

# Test Cases
print(length_of_longest_substring_2("abcabcbb"))  # Output: 3
print(length_of_longest_substring_2("bbbbb"))     # Output: 1
print(length_of_longest_substring_2("pwwekaeew")) # Output: 4
print(length_of_longest_substring_2(""))          # Output: 0
```
