### Task #1: Find First Unique Character in a String

```python
# Solution #1
s = 'loveleetcode'

for i, ch in enumerate(s):
    if s.index(ch) == s.rindex(ch):
        print(s[i])
        break
```

```python
# Solution #2
s = 'loveleetcode'

for i, ch in enumerate(s):
    if ch not in s[i+1:]:
        print(s[i])
        break
```

```python
# Solution #3 (the same complexity as Solution #1)
s = 'loveleetcode'

for ch in s:
    if s.index(ch) == s.rindex(ch):
        print(ch)
        break
```

### Compare Solution #1 vs Solution #2

| Metric                               | Solution 1 (`index/rindex`) | Solution 2 (`slices + not in`) |
| ------------------------------------ | --------------------------- | ------------------------------ |
| Time per iteration                   | O(n) + O(n)                 | O(n) + O(n) + O(n) + O(n)      |

| Best-case Time                       | O(n)                        | O(n)                           |

| Creates temporary strings            | No                          | Yes                            |
| Number of string scans per iteration | 2                           | Up to 4                        |
| Readability                          | High                        | Medium                         |
| Performance                          | Faster                      | Slower                         |


### Task #2: Palindrome Check

Given a string s, return True if it is a palindrome, otherwise False. Ignore case.

```python
# Solution #1 - simpliest (just Ignore case)
def is_palindrome(s: str) -> bool:
    s = s.lower()
    return s == s[::-1]

print(is_palindrome("Level"))
```

```python
# Solution #2: with removing all not alphanum chars + lowercase
def is_palindrome_2(s: str) -> bool:
    res = ''.join(ch.lower() for ch in s if ch.isalnum())
    return res == res[::-1]

print(is_palindrome("_Level"))
print(is_palindrome_2("_Level"))
```

### Task #3: Count Character Frequency
Given a string s, return a dictionary mapping character → count.

```python
# Solution #1: manual counting
def count_char_freq(s: str) -> dict:
    freq = {}
    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1
    return freq

print(count_char_freq("hello world"))
```

```python
# Solution #2: by using collection
from collections import Counter

def count_char_freq_2(s: str) -> dict:
    return dict(Counter(s))

print(count_char_freq_2("hello world"))
```

### Performance comparison

### Task #4: Longest Substring Without Repeating Characters

Return the length of the longest substring without repeating characters.

Note: This is a classic sliding window problem.

```python
# Solution #1: using sliding window technique (using Set)
def length_of_longest_substring(s: str) -> int:
    char_set = set()
    left = 0
    max_length = 0

    for right in range(len(s)):
        while s[right] in char_set:
            char_set.remove(s[left])
            left += 1
        char_set.add(s[right])
        max_length = max(max_length, right - left + 1)

    return max_length

# Test Cases
print(length_of_longest_substring("abcabcbb"))  # Output: 3
print(length_of_longest_substring("bbbbb"))     # Output: 1
print(length_of_longest_substring("pwwekaeew")) # Output: 4
print(length_of_longest_substring(""))          # Output: 0
```

```python
# Solution #1a: using sliding window technique (using List)
def length_of_longest_substring(s: str) -> int:
    char_list = []
    left = 0
    max_length = 0

    for right in range(len(s)):
        while s[right] in char_list:
            char_list.remove(s[left])
            left += 1
        char_list.append(s[right])
        max_length = max(max_length, right - left + 1)

    return max_length

# Test Cases
print(length_of_longest_substring("abcabcbb"))  # Output: 3
print(length_of_longest_substring("bbbbb"))     # Output: 1
print(length_of_longest_substring("pwwekaeew")) # Output: 4
print(length_of_longest_substring(""))          # Output: 0
```

### Compare approaches with Set() vs List[]

List makes the algorithm much less efficient.
| Option | Works? | Recommended? | Reason      |
| ------ | ------ | ------------ | ----------- |
| `set`  | ✅ Yes  | ⭐ Best       | O(1) lookup |
| `list` | ✅ Yes  | ❌ Not ideal  | O(n) lookup |


#### ⚠️ Why list is worse

| Operation        | Complexity |
| ---------------- | ---------- |
| operation (`in`) | O(n)       |
| remove()         | O(n)       |
| worst-case       | O(n²)      |


#### 🚀 Why set is preferred

Set gives:
| Operation     | Complexity |
| ------------- | ---------- |
| lookup (`in`) | O(1)       |
| add           | O(1)       |
| remove        | O(1)       |




```python
# Solution #2: using dictionary to store last seen index of characters
def length_of_longest_substring_2(s: str) -> int:
    char_index = {}
    left = 0
    max_length = 0

    for right in range(len(s)):
        if s[right] in char_index and char_index[s[right]] >= left:
            left = char_index[s[right]] + 1
        char_index[s[right]] = right
        max_length = max(max_length, right - left + 1)

    return max_length

# Test Cases
print(length_of_longest_substring_2("abcabcbb"))  # Output: 3
print(length_of_longest_substring_2("bbbbb"))     # Output: 1
print(length_of_longest_substring_2("pwwekaeew")) # Output: 4
print(length_of_longest_substring_2(""))          # Output: 0
```

### Task #5: Convert a list of strings to integers

```python
# Solution #1: Using a list comprehension to filter out non-digit strings and convert the remaining strings to integers.
def convert_to_integers(str_list: list[str]) -> list[int]:
    return [int(s) for s in str_list if s.isdigit() or (s.startswith('-') and s[1:].isdigit())]

# Test cases
print(convert_to_integers(["1", "2", "3"]))  # Output: [1, 2, 3]
print(convert_to_integers(["10", "20", "30"]))  # Output: [10, 20, 30]
print(convert_to_integers(["-1", "N0", "1"]))  # Output: [-1, 0, 1]
```

```python
# Solution #2: Using map() and filter() functions to achieve the same result.
def convert_to_integers_2(str_list: list[str]) -> list[int]:
    def is_valid_integer(s: str) -> bool:
        return s.isdigit() or (s.startswith('-') and s[1:].isdigit())

    valid_strings = filter(is_valid_integer, str_list)
    return list(map(int, valid_strings))

# Test cases
print(convert_to_integers(["1", "2", "3"]))  # Output: [1, 2, 3]
print(convert_to_integers(["10", "20", "30"]))  # Output: [10, 20, 30]
print(convert_to_integers(["-1", "N0", "1"]))  # Output: [-1, 0, 1]
```

```python
# Solution #3: without functions
strings = ["1", "2", "3", "4"]
integers = list(map(int, strings))

print(integers)  # Output: [1, 2, 3, 4]
```

### Task #6: Check if two strings are anagrams

```python
# Solution #1: Sort and compare
def are_anagrams(s1: str, s2: str) -> bool:
    return sorted(s1) == sorted(s2)

# Test Cases
print(are_anagrams("listen", "silent"))  # Output: True
print(are_anagrams("hello", "world"))    # Output: False

# Time Complexity: 
#  - Sorting each string: O(n log n)
#  - Comparing: O(n)

# Total: O(n log n)

# Space Complexity
#  - Sorting creates new arrays: O(n)
```

```python
# Solution #2: Using a frequency dictionary (hash map)
def are_anagrams_2(s1: str, s2: str) -> bool:
    if len(s1) != len(s2):
        return False
 
    count = {}  # eg. {'l': 2, 'i': 1, 's': 1, 't': 1, 'e': 1, 'n': 1}

    for ch in s1:
        count[ch] = count.get(ch, 0) + 1

    for ch in s2:
        if ch not in count or count[ch] == 0:
            return False
        count[ch] -= 1

    return True

# Test Cases
print(are_anagrams_2("listen", "silent"))  # Output: True
print(are_anagrams_2("hello", "world"))    # Output: False

# Time Complexity
#  - First loop over s1: O(n)
#  - Second loop over s2: O(n)

# Total: O(n)

# Space Complexity
#  - Frequency dictionary: O(k) (unique characters)
```
