## Mutable Default Arguments

In Python, **default arguments are evaluated only once when the function is defined**, not each time the function is called. This means that if you use a **mutable object** (like a list, dictionary, or set) as a default argument, **all future calls to that function will share and mutate the exact same object**.

### The Problem

When you define a function with a mutable default argument, Python binds that object to the function definition. If you modify that object inside the function, the modifications will persist into subsequent calls

### Buggy Code Example

```python
def add_item(item, item_list=[]):
    item_list.append(item)
    return item_list

print(add_item("apple"))  # Output: ['apple']
print(add_item("banana")) # Output: ['apple', 'banana'] (Expected: ['banana'])
print(add_item("cherry")) # Output: ['apple', 'banana', 'cherry']

# Desc: The issue here is that the default value for item_list is mutable (a list), 
# and it is shared across all calls to the function. 
# To fix this, we should use None as the default value and create a new list inside the function if item_list is None.
```

### The Solution

The standard Pythonic solution is to **use** *None* **as a sentinel value** for the default argument. You then explicitly check if the argument is *None* inside the function body and initialize a new mutable object at that time.

### Corrected Code Example

```python
def add_item(item, item_list=None):
    if item_list is None:
        item_list = []  # Creates a fresh, new list for this specific call
    item_list.append(item)
    return item_list

print(add_item("apple"))  # Output: ['apple']
print(add_item("banana")) # Output: ['banana']
print(add_item("cherry")) # Output: ['cherry']
```

## NOTE: Default arguments are evaluated once at function definition time, not every time the function is called.

If you use this in an Airflow task or a long-running worker, your batch will persist data across different pipeline runs, causing state pollution and memory leaks.
