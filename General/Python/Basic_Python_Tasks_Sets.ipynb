## Sets

13. Remove duplicates from a list

Description:
Given a list of numbers, return a list with duplicates removed.

```python
def remove_duplicates(nums):
    return list(set(nums))

if __name__ == "__main__":
    sample_list = [1, 2, 2, 3, 4, 4, 4]
    print("Original list:", sample_list)
    print("List after removing duplicates:", remove_duplicates(sample_list))

# Output: [1, 2, 3, 4] (order may vary)
```

14. Find common elements between two lists

Description:
Return items that appear in both lists.

```python
def intersection(a, b):
    return list(set(a) & set(b))

print(intersection([1, 2, 3], [2, 3, 4]))  
# Output: [2, 3] (order may vary)
```

15. Check if a string contains all unique characters

Description:
Return True if all characters in a string are unique.

```python
def all_unique(s):
    return len(set(s)) == len(s)

print(all_unique("abcde"))   # True
print(all_unique("hello"))   # False (because: l repeats)
```
