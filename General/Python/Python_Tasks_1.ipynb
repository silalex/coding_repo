# Strings
### 1. Find the first unique character in a string

```python
# Find 1st unique char in a string & return this 'char'
def find_first_unique(s: str) -> str:
    for c in s:
        if s.index(c) == s.rindex(c):
            return c
        
# Test case
s = 'leetcode with llama'
find_first_unique(s) # Output: 'c'
```

```python
# Find 1st unique char in a string & return 'char' index
def find_first_unique(s: str) -> int:
    for i, c in enumerate(s):
        if s.index(c) == s.rindex(c):
            return i
        
# Test case
s = 'leetcode with llama'
find_first_unique(s) # Output: 4

# NOTE: List Comprehension can't be used because it generates a new list()
```

### 2. Remove duplicated numbers from the sorted string
In this task we're going to use map()

```python
# By using set() - 1 single line of code
def remove_duples(s: str) -> list[int]:
    return list(set(map(int, s.split(','))))

# Test case
s = '1,1,2,2,2,3,4,4,5'
remove_duples(s)
```

```python
# By using FOR loop - manual checks
def remove_duples(s: str) -> list[int]:
    nums = list(map(int, s.split(',')))
    res = [nums[0]]

    # NOTE: List Comprehension can't be used because we need to check the last element in the list
    for num in nums[1:]:
        if num != res[-1]:
            res.append(num)

    return res

# Test case
s = '1,1,2,2,2,3,4,4,5'
remove_duples(s)
```

# Lists

### 1. Missed number in the List with sorted numbers

```python
# Gauss formula
def find_missed_num(nums: list[int]) -> int:
    n = len(nums)
    exepcted_sum = n * (n + 1) // 2  # "ideal" List sum
    current_sum = sum(nums)

    return exepcted_sum - current_sum

# Test case
nums = [6,0,1,4,2,3]
print(find_missed_num(nums))
```

# Dicts (Maps / Hash tables)
### 1. Count Word Frequency in ta LIST into DICT

```python
def words_cnt(words: list[str]) -> dict[str: int]:
    res = {}

    for word in words:
        res[word] = res.get(word, 0) + 1

    return res

# Test case
print(words_cnt(["apple", "banana", "apple", "orange", "banana", "apple"]))
```

# map()

### 1. Tuples concatination vs map() usage

```python
# Simple concationation of two tuples
print(('apple', 'banana', 'cherry') + ('orange', 'lemon', 'pineapple'))
```

```python
# Map a function over two tuples: to concatenate tuple elements
def myfunc(a, b):
  return a + b

x = map(myfunc, ('apple', 'banana', 'cherry'), ('orange', 'lemon', 'pineapple'))

#convert the map into a list, for readability:
print(list(x))
```

### 2. Calculate the length of each word in a given tuple

```python
def word_len(s: str) -> int:
    return len(s)

x = tuple(map(word_len, ('orange', 'lemon', 'pineapple')))
print(x)
```
