## Task: 3Sum Zero Indexes (Return all IDXs where nums[i] + nums[j] + nums[k] = 0)
```
You are given an integer array nums.

Return all unique triplets of indices (i, j, k) such that: i, j, and k are distinct indices

nums[i] + nums[j] + nums[k] = 0

The order of indices inside a triplet does not matter

The order of triplets in the result does not matter
```

````python
from typing import List, Tuple

def three_sum_zero_indexes(nums: List[int]) -> List[Tuple[int, int, int]]:
    n = len(nums)
    res = set()

    # (value, original_index)
    arr = sorted((v, i) for i, v in enumerate(nums))

    for i in range(n - 2):
        left = i + 1
        right = n - 1

        while left < right:
            s = arr[i][0] + arr[left][0] + arr[right][0]

            if s == 0:
                res.add(tuple(sorted([
                    arr[i][1],
                    arr[left][1],
                    arr[right][1]
                ])))
                left += 1
                right -= 1

            elif s < 0:
                left += 1
            else:
                right -= 1

    return sorted(list(res))

# Example usage
nums = [-1, 0, 1, 2, -1, -4]
print(three_sum_zero_indexes(nums)) # Output: [(0, 1, 2), (0, 4, 3)] or similar triplets with original indices
````
