## 1. Index of the First Unique Character in a String
```
Task: Given a string s, return the index of the first non-repeating character.
If it does not exist, return -1.

Source: LeetCode — 387. First Unique Character in a String

Example:
Input:  "leetcode"
Output: 0
```

````python
def return_first_unique_char_index(s: str) -> int:
    for c in s:
        if s.count(c) == 1:
            return s.index(c)
        
    return -1


# Example usage:
if __name__ == "__main__":
    test_string = "Hello Heppy World!"
    print(return_first_unique_char_index(test_string))  # 10
````

## 2. Reverse a String
```
Task: Given a string s, return a new string with characters reversed.

Source: LeetCode — 344. Reverse String

Example:
Input:  "wolf"
Output: "flow"
```

````python
def string_reverse(s: str) -> str:
    return s[::-1]

# Example usage:
if __name__ == "__main__":
    test_string = "Hello Heppy World!"
    print(string_reverse(test_string))  # !dlroW yppeH olleH
````

## 3. Palindrome Check
```
Task
Return True if the string is a palindrome (ignore case and non-alphanumeric characters).

Source: LeetCode — 125. Valid Palindrome

Example:
Input:  "A man, a plan, a canal: Panama"
Output: True


A phrase is a palindrome if, after converting all uppercase letters into lowercase letters and removing all non-alphanumeric characters, it reads the same forward and backward. Alphanumeric characters include letters and numbers.

Given a string s, return true if it is a palindrome, or false otherwise.

Example 1:
Input: s = "A man, a plan, a canal: Panama"
Output: true
Explanation: "amanaplanacanalpanama" is a palindrome.

Example 2:
Input: s = "race a car"
Output: false
Explanation: "raceacar" is not a palindrome.

Example 3:
Input: s = " "
Output: true
Explanation: s is an empty string "" after removing non-alphanumeric characters.

Since an empty string reads the same forward and backward, it is a palindrome.
```

````python
def string_palindrome_check(s: str) -> bool:
    # remove all not alphanumeric characters and convert to lowercase
    res = ''.join(c.lower() for c in s if c.isalnum())
    return res == res[::-1]

# Example usage:
if __name__ == "__main__":
    test_string = "A man, a plan, a canal: Panama" # "amanaplanacanalpanama"
    print(string_palindrome_check(test_string))    # True
````

## 4. Count Character Frequency
```
Task: Return a dictionary mapping each character to its count.

Source: HackerRank — Collections.Counter()

Example
Input:  "aabbc"
Output: {'a': 2, 'b': 2, 'c': 1}
```

````python
# Task: Return a dictionary mapping each character to its count.
# use Collections.Counter()

from collections import Counter

def return_char_count_dict(s: str) -> dict:
    return dict(Counter(s))

# Example usage:
if __name__ == "__main__":
    test_string = "Hello Heppy World!"
    print(return_char_count_dict(test_string))  # {'H': 2, 'e': 2, 'l': 3, 'o': 2, ' ': 2, 'p': 2, 'y': 1, 'W': 1, 'r': 1, 'd': 1, '!': 1}
````

## 6. Longest Substring Without Repeating Characters
```
Task: Return the length of the longest substring without repeating characters.

Source: LeetCode — 3. Longest Substring Without Repeating Characters

Example 0:
Input:  "abcabcbb"
Output: 3

Example 1:
Input: s = "abcabcbb"
Output: 3
Explanation: The answer is "abc", with the length of 3. Note that "bca" and "cab" are also correct answers.

Example 2:
Input: s = "bbbbb"
Output: 1
Explanation: The answer is "b", with the length of 1.

Example 3:
Input: s = "pwwkew"
Output: 3
Explanation: The answer is "wke", with the length of 3.

Notice that the answer must be a substring, "pwke" is a subsequence and not a substring.
```

````python
def return_the_longest_nonrepeatable_substring_len(s: str) -> int:
    last_seen = {}      # Dictionary to store the last index of each character
    left = 0            # Left pointer of the sliding window
    max_len = 0         # Maximum length of non-repeating substring found

    for right, char in enumerate(s):
        if char in last_seen and last_seen[char] >= left:
            left = last_seen[char] + 1

        last_seen[char] = right
        max_len = max(max_len, right - left + 1)

    return max_len

# Example usage:
if __name__ == "__main__":
    test_string = "abcavwxyzbcbb"
    print(return_the_longest_nonrepeatable_substring_len(test_string))  # 8 (avwxyzbc)
````

## 7. Two Sum
```
Task: Return indices of two numbers that add up to target.
Source: LeetCode — 1. Two Sum

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

Example 1:
Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

Example 2:
Input: nums = [3,2,4], target = 6
Output: [1,2]

Example 3:
Input: nums = [3,3], target = 6
Output: [0,1]
```

````python
# Solution 1: O(n²) complexity
 
from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for i in range(len(nums)):
            for j in range (i + 1, len(nums)):
                if nums[i] + nums[j] == target:
                    return [i, j]

        # solution didn't find
        return []
````

````python
# Solution 2: O(n) complexity

from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        seen = {}  # value -> index

        for i, num in enumerate(nums):
            complement = target - num

            if complement in seen:
                return [seen[complement], i]

            seen[num] = i
````

````python
# Example usage:
if __name__ == "__main__":
    sol = Solution()
    print(sol.twoSum([15,7,11,2], 9))  # Expected output: [1, 3]
````
