# 1. Index of the First Unique Character in a String
```
Given a string s, return the index of the first non-repeating character.
If it does not exist, return -1.

Source: LeetCode â€” 387. First Unique Character in a String

Example:
Input:  "leetcode"
Output: 0
```

````python
def return_first_unique_char_index(s: str):
    for c in s:
        if s.count(c) == 1:
            return s.index(c)
        
    return -1


# Example usage:
if __name__ == "__main__":
    test_string = "Hello Heppy World!"
    print(return_first_unique_char_index(test_string))  # 10
````
